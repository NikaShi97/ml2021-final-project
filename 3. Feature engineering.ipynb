```python
import pathlib
import numpy as np
import pandas as pd

data = pathlib.Path('/home/ivan/PycharmProjects/ml2021-final-project/data')
```

```python
df = pd.read_csv(data / 'undersampled_shuffled.csv')
```

```python
df.sample(7)
```

```python
df['NDVI'] = (df.B8 - df.B4) / (df.B8 + df.B4)
df['mNDVI'] = (df.B8 - df.B4) / (df.B8 + df.B4 - 2 * df.B1)
df['NGRDI'] = (df.B3 - df.B5) / (df.B3 + df.B5)
df['DVI'] = df.B8 - df.B4
df['S2RVI'] = df.B8 / df.B4
df['GNDVI'] = (df.B8 - df.B3) / (df.B8 + df.B3)
df['IRECI'] = (df.B7 - df.B4) / (df.B5 / df.B6)
df['NDWI'] = (df.B3 - df.B8) / (df.B3 + df.B8)
df['Rededge1'] = df.B5 / df.B4
df['Rededge2'] = (df.B5 - df.B4) / (df.B5 + df.B4)
df['RBNDVI'] = (df.B8 - df.B5 + df.B1) / (df.B8 + df.B5 + df.B1)
df['GRNDVI'] = (df.B8 - df.B3 + df.B1) / (df.B8 + df.B3 + df.B1)
df['GBNDVI'] = (df.B8 - df.B3 + df.B5) / (df.B8 + df.B3 + df.B5)

df['RFDI'] = (df.Gamma0_VV - df.Gamma0_VH) / (df.Gamma0_VV + df.Gamma0_VH)
df['S1RVI'] = 4 * df.Gamma0_VH / (df.Gamma0_VV + 2 * df.Gamma0_VH)
df['Gamma0_VV_db'] = 10 * np.log10(abs(df.Gamma0_VV))
df['Gamma0_VH_db'] = 10 * np.log10(abs(df.Gamma0_VH))
df['Gamma0_Ratio_db'] = 10 * np.log10(abs(df.Gamma0_Ratio))

df.drop(['Gamma0_VV', 'Gamma0_VH', 'Gamma0_Ratio'], axis=1, inplace=True)
```
